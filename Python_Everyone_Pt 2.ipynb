```python
# Dictionaries & Files

# A common use of a dict is to count the occurrence of words in a file with text.
# Python - read thru the of the file, break each line into a list of words and then loop thru each word in the
# line and count each word using a dict.

# Note 2 for loops. Outer lopp is reading the lines of the file.Ineer loop is iterating thru each of the word on
# that particular line (nested loops). the 2 nested loops ensure that every word on every line is counted.

# else statement - more compact incrementing a variable. 'counts[word] += 1' is equivalent to 'counts[word] =
#  counts[words] + 1'

# Output - raw dump of all counts in unsoreted hash order.



fname = 'mbox-short.txt'
try:
    fhand = open(fname)
except:
    print('Cannot open file: ', fname)
    exit()
       
       
counts = dict()
for line in fhand:
       words = line.split()
       for word in words:
           if word not in counts:
               counts[word] =1
           else:
               counts[word] += 1
print(counts)
```

```python
#Advance text parsing

# Using string methods - lower, punctuation, and translate. Translate is the most subtle
# line.translate(str.maketrans(fromstr, tostr, deletestr())

import string

fname = 'mbox-short.txt'
try:   
    fhand = open(fname)
except:
    print('Cannot open file: ', fname)
    exit()
    
counts = dict()
for line in fhand:
    line = line.rstrip()
    line = line.translate(line.maketrans('', '', string.punctuation))
    line = line.lower()
    words = line.split()
    for word in words:
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1
            
print(counts) 

```

```python
# Sort from longest to shortest 

# First loop builds a list of tuples, where each tuple is a word preceded by the length
# sort compares the first element, length, first n only consider the 2nd element to break ties
# arg (reverse=True) tells sort to go in descending order
# Second loop traverses the list of tuples and buis a list of words in descending order of length

# This is a DSU pattern/code (Decorate,Sort Undecorate)

txt = 'but soft what light in yonder window breaks'
words = txt.split()
t = list()
for word in words:
    t.append((len(word), word))
    
t.sort(reverse=True)

res = list()
for length, word in t:
    res.append(word)
    
print(res)
    
```

```python
# Regular Expressions

# RE - a powerful library but complicated. Like own programming language for searching/parsing strings
# RE library 're' must be imported into program b4 use

# Open the file, loop thru each line n use RE search() to only print out lines that contain the string 'From'
import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('From:', line):
        print(line)
```

```python
# Power of RE comes when we add special character to the search string that allow more precise control
#  which lines matach the string
# The caret in RE match is used to match the beginning of a line

import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:', line):
        print(line)
```

```python
# RE - Character matching

# Special Character - 'period'

# RE 'F..m:' would match any strings 'From:', 'F12m:', 'F!@m:' since the 'period' in the RE match any character

# Search fo lines that start with 'F', followed by 2 characters, followed by 'm'

import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From..m:', line):
        print(line)
```

```python
# RE - ability to indicate that a character can be repeated any number of times using '*' or '+'

# Search string 'From:.+@' match lines that start with 'From:' followed by one or more characters (.+),
#  followed by @ sign

# Instead of matching a single character in the search-string, they match zero or more characters (the asterik)
#  or one or more character (the plus sign)

# Think (.+) as a wildcard expanding to match all characters btw the colon and the @ sign

import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:.+@', line):
        print(line)
```

```python
# RE = Extracting data

# extract date from a string; uses 'findall()'

# Pull the email addresses from each of the lines.

# 'findall()' method searches the string in the second arg and returns a list of all strings that look like email
#  addresses. Using a 2 character sequence that matches a non-whitespace character (\S)

# Looking for substrings that have at least one non whitespace character followed by @ sign, followed by at least
#  one more non whitespace character

# RE would not match the string '@11PM' because there are no non blank characters b4 the @ sign

import re
s = 'A message from ulysses@umich.ed to ming@iupui.edu about fun @11PM'
lst = re.findall('\S+@\S+', s)
print(lst)
```

```python
# Search for lines that hv @ sign btw characters

# findall() retrurns a list. Check if the number of elements in out returnded list is more than zero to print
#  only lines where at least one substring looks like an email address

# '\S' is asking to match the set of non whitespace character


import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('\S+@\S+', line)
    if len(x) > 0:
        print(x)
    
```

```python
# Get portion of strings that starts and ends with a letter or number

# Square brackets to indicate a set of multiple acceptable characters to consider matching.
# The \S is asking to match the set of non whitespace characters
# Looking for substrings that start with a single lowercase letter, uppercase letter or number '[a-zA-Z0-9]'
#  followed by zero or more non blank characters (\S*), followed by an uppercase or lowercase letter.
# Switch from + to * to indicate zero or more non blank characters since [a-zA-Z0-9] is one non blank 
#  character. The * or + applies to the single character immediately to the left of the plus or asterik


import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('[a-zA-Z0-9]\S*@\S*[a-zA-Z]', line)
    if len(x) > 0:
        print(x)
```

```python
# Combining searching n extracting

# Search for lines that start with 'X' followed by any non whitespace character and ':' followed by a space and
#  any number; the number can include a decimal 

import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()  
    if re.search('X\S*: [0-9.]+', line):
        print(line)
```

```python
# Search for lines that start with 'X' followed by any non whitespace character and ':' followed by a space and
#  any number; the number can include a decimal. Print the number if greater than zero

# Instead of calling search(), add parentheses around the part of the RE that represents the floating point-number
#  to indicat that we only wnat findall() to give the gloating point number of the matching string


import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()  
    x = re.findall('X\S*: ([0-9.]+)', line)
    if len(x) > 0:
        print(x)
```

```python
# Search for lines that start with 'Details:rev=', followed by numbers and '.'. Then print the number if it
# is greater than zero

# Note [0-9] is greedy and it tries to make as large a string of digits before extracting those digits.

import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()  
    x = re.findall('Details:.*rev=([0-9.]+)', line)
    if len(x) > 0:
        print(x)
```

```python
# Search for lines that start with From and a character followed by a 2 digit number btw 00 n 99 followed by
#  by ':', then print the number if greter than zero

# Lo0king for lines that start with From(note the space), followed by any number of characters(.*), followed
#  by a space, followed by 2 digits [0-9][0-9], followed by a colon

# In order to pull out only the hour using finall(), add parentheses around the 2 digits ([0-9][0-9]):

import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip() 
    x = re.findall('From .*([0-9][0-9]):',line)
    if len(x)>0:print(x)
        
```

```python
# Escape character

# We use special characters in RE to match the beg or end of line or specify wild cards, we need to indicate that
# these characters are 'normal' n we want to match the actual character such as $ sign ir caret. We can indicate 
# that we want to simply match a char by prefixing that char with a backlash. Since we prefix the $ sign with a
# backlash, it actually matches the $ sign in the input string instead of matching the 'end of line', and the one
# or more digits or the period char. Note: Inside square brackets, char are not special. Outside square brackets, a
# period is the 'wild card' n matches any char. Inside the square brackets, the period is a period.

import re
x = 'we just received $10.00 for cookies'
y = re.findall('\$[0-9.]+',x)
print(y)
```

```python
# Summary

# RE are search strings with special char that communicate to the RE as what defines 'matching' n what is extracted
# from the matched strings

# ^  matches the beg of the line
# $  matches the end of the line
# .  matches any char ( a wild card)
# \s matches a whitespace char
#\S  matches a non whitespace char
# *  applies to the immediate preceding char n indicate to match zero or more times
#*?  applies to the immediate preceding char n indicate to match zero or more times in non greedy mode
# +  applies to the immediately preceding char n indicates to match one or more times
# +? applies to the immediatlely preceding char n indicates to match one or more times in non greedy mode
# ?  applies to the immediately preceding char n indicated to match zero or one time
#??  applies to the immediately preceding character n indicates to match zerp or one time in non greedy mode
#[aeiou] matches a single character as long as the char is in the specified set
#[a-z0-9] specify ranges of char using the minus sign. Eg, this is a single char that must be a lowercase letter
#         or digit
# [^A-Za-z] when the first char in the set is a caret, it inverts the logic. Eg. Matches a single char that is
#           anything other than an uppercase or lowercase letter
# () when parentheses are added to RE, they are ignored for the purpose of matching, but allow you to extract 
#    a particular subset of the matched string when using findall()
#\b matches the empty string but only at the start or end of a word
#\B matches the empty string, but not at the start or end of a word
#\d matches any decimal digit; equivalent to the set[0-9]
#\D matches any non digit char; equivalent to the set[^0-9]
```
